```python
from bs4 import BeautifulSoup
import requests

url = 'https://nupes-2022.fr/le-programme/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
li = soup.find('main', class_='elementor-column elementor-col-100 elementor-top-column elementor-element elementor-element-3453020').find('div', class_='elementor-widget-wrap elementor-element-populated')
```

```python
programme = [u.text.replace('\n', '') for u in li.find_all('div') if 'data-id' in u.attrs and u.text.replace('\n', '') != '']
```

```python
keys = programme[5::2]
```

```python
value = programme[6::2]
```
